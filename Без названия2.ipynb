```python
col = ['text']
test_pos = pd.read_csv("test_pos.csv", header = None, names = col)
```

```python
col = ['text']
test_neg = pd.read_csv("test_neg.csv", header = None, names = col)
```

```python
import re

from pymorphy2 import MorphAnalyzer
from nltk.corpus import stopwords

patterns = "[A-Za-z0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"
stopwords_ru = stopwords.words("russian")
morph = MorphAnalyzer()

def lemmatize(doc):
    doc = re.sub(patterns, ' ', doc)
    tokens = ''
    for token in doc.split():
        if token and token not in stopwords_ru:
            token = token.strip()
            token = morph.normal_forms(token)[0]
            
            tokens += token
            tokens += ' '
    if len(tokens) > 2:
        return tokens
    return None
```

```python
data_pos = test_pos['text'].apply(lemmatize)
data_pos
```

```python
data_neg = test_neg['text'].apply(lemmatize)
data_neg
```

```python
sp_pos = []
for i in range(len(test_pos)):
    sp_pos.append(0)
data_pos = pd.DataFrame(data_pos)
data_pos['key'] = sp_pos
data_pos
```

```python
sp_neg = []
for i in range(len(data_neg)):
    sp_neg.append(1)
data_neg = pd.DataFrame(data_neg)
data_neg['key'] = sp_neg
data_neg
```

```python
data = data_pos.append(data_neg)
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
data = data.dropna()
data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```
